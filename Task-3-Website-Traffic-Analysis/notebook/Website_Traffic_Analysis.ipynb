```python
top_countries = df['country'].value_counts().head(10)

print(top_countries)
```

```python
plt.figure(figsize=(10,5))

sns.barplot(
    x=top_countries.values,
    y=top_countries.index
)

plt.title("Top Countries")

plt.xlabel("Clicks")

plt.ylabel("Country")

plt.show()
```

```python
top_cities = df['city'].value_counts().head(10)

print(top_cities)

plt.figure(figsize=(10,5))

sns.barplot(
    x=top_cities.values,
    y=top_cities.index
)

plt.title("Top Cities by Activity")

plt.xlabel("Number of Interactions")

plt.ylabel("City")

plt.show()
```

```python
daily_activity = df.groupby(df['date'].dt.date).size()
```

```python
plt.figure(figsize=(12,5))

daily_activity.plot()

plt.title("Daily Traffic Trend")

plt.xlabel("Date")

plt.ylabel("Events")

plt.show()
```

```python
sessions = df['linkid'].nunique()

print("Total Sessions:", sessions)
```

```python
users = df[['city', 'country']].drop_duplicates().shape[0]

print("Estimated Unique Users:", users)
```

```python
bounce_sessions = df.groupby('linkid').size()

bounce_rate = (
    bounce_sessions[bounce_sessions == 1].count()
    / len(bounce_sessions)
) * 100

print("Bounce Rate:", round(bounce_rate, 2), "%")
```

Average session duration could not be calculated because the dataset did not contain timestamp intervals or session duration fields.

```python
top_links = df['linkid'].value_counts().head(10)

print(top_links)
```

```python
plt.figure(figsize=(12,5))

sns.barplot(
    x=top_links.values,
    y=top_links.index
)

plt.title("Top Accessed Links")

plt.xlabel("Interactions")

plt.ylabel("Link ID")

plt.show()
```

```python
country_activity = df.groupby('country').size().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))

sns.barplot(
    x=country_activity.values,
    y=country_activity.index
)

plt.title("Top User Countries")

plt.xlabel("Activity Count")

plt.ylabel("Country")

plt.show()
```

```python
daily_activity = df.groupby(df['date'].dt.date).size()

plt.figure(figsize=(12,5))

daily_activity.plot()

plt.title("Daily Website Traffic Trend")

plt.xlabel("Date")

plt.ylabel("Events")

plt.show()
```

```python

top_tracks = df['track'].value_counts().head(10)

print(top_tracks)

plt.figure(figsize=(10,5))

sns.barplot(
    x=top_tracks.values,
    y=top_tracks.index
)

plt.title("Popular Tracks")

plt.xlabel("Number of Interactions")

plt.ylabel("Track")

plt.show()
```

1. Certain countries generated significantly higher website interactions than others.

2. Some links received substantially more traffic, indicating high user interest.

3. Event activity varied across different dates, showing traffic fluctuation trends.

4. A notable percentage of sessions contained only one interaction, indicating possible bounce behavior.

5. Popular artists and tracks attracted the highest engagement levels.

1. Improve engagement on low-performing links with better content and CTA placement.

2. Focus advertising campaigns on high-traffic countries and cities.

3. Promote highly engaging artists/tracks more prominently.

4. Optimize user experience on pages with high bounce behavior.

5. Analyze peak traffic periods to improve marketing and promotional strategies.

Dataset Limitation Note:

The dataset did not contain traditional web analytics fields such as session IDs, user IDs, landing pages, exit pages, or session duration. Therefore, sessions, users, and bounce rates were approximated using available interaction and link data. User flow analysis was adapted using top accessed links and interaction trends.

This project analyzed website interaction and traffic activity data to understand user engagement patterns, geographic activity, and content popularity. Data cleaning and preprocessing techniques were applied using Python libraries such as Pandas, Matplotlib, and Seaborn.
